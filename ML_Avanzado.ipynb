## Topologia de Red
Describe el numero de neuronas y capas, asi como la manera en que se conectan.

- La topologia habla de como acomodamos nuestra red basados en nuestras 3 caracteristicas de construccion.
  1. Numero de capas(Layers). Se dividen en receptores(input), resultados(output). Cada capa de la red incluye un termino de sesgo.
  2. Numero de Nodos, El numero de nodos de resultados depende del tipo de tarea. como regla asociada a el numero de neuronas se usa.

  $n_0 = C\sqrt{nm}$, Donde C es una constante (4 por lo general), $n$ es el numero de neuronas de entrada y $m$ es el numero de neuronas de salida.

Para su entrenamiento las redes usan iteraciones.

- Primero comienza asignando unos valores aleatorios a ($w$).
- En el proceso *FORWARD* se activan las neuronas aplicando el peso a los input hasta alcanzar la funcion de activacion que permite generar los valores de salida.
- Se testea el error con los nodos de salida *LOG-LOSS*, es decir cuantificamos el error y nos preparamos para ajustar los pesos.
- Se realiza el back-propagation donde se corrigen los pesos($w$). el proceso se conoce como *BACKWARD*

Nuestro criterio de optimizacion es minimizar el error por lo que.

$$w_*i = w_1 - \gamma \frac{\partial \text{err_fun}}{\partial w_i}$$

# Ensamble

Los Metodos de ensambles nacen de la idea de unir nuestros modelos con bajos desempenio para lograr mejorar las predicciones de estos.

Para esto se usan unas funciones de asignacion(que deternima el entrenamiento) y unas de combinacion(que genera el output)

- Para la asignacion, variamos los datos de entrada de cada modelo. (ya sean muestreos tipo bootstrap o diferentes modelos).
- Para la combinacion, usamos promedios simples o ponderaciones de las predicciones resultantes.

El Bootstrap es una tecnica de muestreo con reemplazo, es decir de la misma muestra genero varia submuestras las cuales pueden tener valores repetidos.

Nos ayuda a generar muestras aleatorias y generalizar la estimacion final.

image.png

## Bagging

El bagging es el metodo mas utilizado de ensamble, simplemente se usa un muestreo tipo bootstrap, luego se usa un unico algoritmo de entrenamiento y por ultimo se combinan las predicciones usando la regla de mayoria o promediando pronosticos.

Bagging nos ayuda a mejorar la estabilidad y Accuracy de nuestros modelos de machine learning

Suele tener mucho mas exito si nuestro modelo genera un sobre entrenamiento. (alta varianza).

El mas popular es el estimador de Bagging Trees o Random Forest.

## Boosting

En el caso del Boosting, se usan modelos en secuencia usando las predicciones de salida como las de entrada en el proximo, el ultimo proceso nos genera el promedio con pesos el cual nos da el resultado.

La base del proceso es hacer un boost a cada modelo ayudando a disminuir el error de cada modelo que va prediciendo.

A diferencia del Bagging entrenamos los modelos en secuencia usando una version con peso de nuestros datos. Esto ayuda a modelos mas sesgados.

Usamos pesos altos en los mal clasificados de cada modelo para crear un enfasis en los casos dificiles.

Un ejemplo es el AdaBoost, Gradient boosting machine (GBM).
